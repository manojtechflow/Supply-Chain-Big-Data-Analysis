```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
supply_dataset = pd.read_csv('./Datasets/preprocess_supply_dataset.csv',encoding='ISO-8859-1', header=0)
```

```python
supply_dataset.head()
```

Creating Dimension Tabels For each Categorical Columns

Data Modeling

```python
# Creating Dimension table for Delivery Status
DeliveryStatusDim = supply_dataset['Delivery Status'].unique()
DeliveryStatusDim = pd.DataFrame(DeliveryStatusDim, columns=['Delivery Status'])
DeliveryStatusDim['Del_Status_Id'] = DeliveryStatusDim.index
DeliveryStatusDim
```

```python
# creating Dimension table for Shipping Mode
ShippingModeDim = supply_dataset['Shipping Mode'].unique()
ShippingModeDim = pd.DataFrame(ShippingModeDim, columns=['Shipping Mode'])
ShippingModeDim['Ship_Mode_id'] = ShippingModeDim.index
ShippingModeDim
```

```python
# creating Dimension table for Customer Country
CustomerCountryDim = supply_dataset['Customer Country'].unique()
CustomerCountryDim = pd.DataFrame(CustomerCountryDim, columns = ['Customer Country'])
CustomerCountryDim['Cust_Country_Id'] = CustomerCountryDim.index 
CustomerCountryDim
```

```python
# Creating Dimension table for Customer City
CustomerCityDim = supply_dataset['Customer City'].unique()
CustomerCityDim = pd.DataFrame(CustomerCityDim, columns= ['Customer City'])
CustomerCityDim['Cust_City_Id'] = CustomerCityDim.index
CustomerCityDim
```

```python
# Creating Dimension table for Customer Segment
CustomerSegmentDim = supply_dataset['Customer Segment'].unique()
CustomerSegmentDim = pd.DataFrame(CustomerSegmentDim, columns=['Customer Segment'])
CustomerSegmentDim['Cust_Seg_Id'] = CustomerSegmentDim.index
CustomerSegmentDim
```

```python
# Creating Dimension table for Payment Type
PaymentTypeDim = supply_dataset['Type'].unique()
PaymentTypeDim = pd.DataFrame(PaymentTypeDim, columns = ['Type'])
PaymentTypeDim['Pay_Type_Id'] = PaymentTypeDim.index
PaymentTypeDim
```

```python
# Creating Dimension table for Department Name
DepartmentDim = supply_dataset['Department Name'].unique()
DepartmentDim = pd.DataFrame(DepartmentDim, columns = ['Department Name'])
DepartmentDim['Dep_Id'] = DepartmentDim.index
DepartmentDim
```

```python
# Creating Dimension table for Market
MarketDim = supply_dataset['Market'].unique()
MarketDim = pd.DataFrame(MarketDim, columns=['Market'])
MarketDim['Market_Id'] = MarketDim.index
MarketDim
```

```python
# Creating Dimension Table for Order City
OrderCityDim = supply_dataset['Order City'].unique()
OrderCityDim = pd.DataFrame(OrderCityDim, columns=['Order City'])
OrderCityDim['Ord_Cty_Id'] = OrderCityDim.index
OrderCityDim
```

```python
# Creating Dimension Table for Order Country
OrderCountryDim = supply_dataset['Order Country'].unique()
OrderCountryDim = pd.DataFrame(OrderCountryDim, columns=['Order Country'])
OrderCountryDim['Ord_Ctry_Id'] = OrderCountryDim.index
OrderCountryDim
```

```python
# Creating Dimesion Table for Category Name

CategoryNameDim = supply_dataset['Category Name'].unique()
CategoryNameDim = pd.DataFrame(CategoryNameDim, columns=['Category Name'])
CategoryNameDim['Ctg_Name_Id'] = CategoryNameDim.index
CategoryNameDim
```

```python
# Creating Fact Table 

fact_table = pd.merge(supply_dataset, MarketDim, on = 'Market', how = 'inner')
```

```python
fact_table = pd.merge(fact_table, DeliveryStatusDim, on = 'Delivery Status', how = 'inner' )
```

```python
fact_table = pd.merge(fact_table, ShippingModeDim, on = 'Shipping Mode', how = 'inner')
```

```python
fact_table = pd.merge(fact_table, CustomerCountryDim, on = 'Customer Country', how = 'inner')
```

```python
fact_table = pd.merge(fact_table, CustomerCityDim, on = 'Customer City', how = 'inner')
```

```python
fact_table = pd.merge(fact_table, CustomerSegmentDim, on= 'Customer Segment', how = 'inner')
```

```python
fact_table = pd.merge(fact_table, PaymentTypeDim, on='Type', how = 'inner')
```

```python
fact_table = pd.merge(fact_table, DepartmentDim, on='Department Name', how = 'inner')
```

```python
fact_table = pd.merge(fact_table, OrderCityDim, on='Order City', how='inner')
```

```python
fact_table = pd.merge(fact_table, OrderCountryDim, on = 'Order Country', how='inner')
```

```python
fact_table = pd.merge(fact_table, CategoryNameDim, on='Category Name', how='inner')
```

```python
# Extrating Year, Month, Day from date for further Analysis
fact_table['order date (DateOrders)'] =pd.to_datetime(fact_table['order date (DateOrders)'])
fact_table['Order Year']= fact_table['order date (DateOrders)'].dt.year
fact_table['Order Month']= fact_table['order date (DateOrders)'].dt.month
fact_table['Order Day']= fact_table['order date (DateOrders)'].dt.weekday
```

```python
fact_table['shipping date (DateOrders)'] =pd.to_datetime(fact_table['shipping date (DateOrders)'])
fact_table['Shipping Year']= fact_table['shipping date (DateOrders)'].dt.year
fact_table['Shipping Month']= fact_table['shipping date (DateOrders)'].dt.month
fact_table['Shipping Day']= fact_table['shipping date (DateOrders)'].dt.weekday
```

```python
# Creating Dimension Table for Order Month 
OrderMonthDim = fact_table['Order Month'].unique()
OrderMonthDim = pd.DataFrame(OrderMonthDim, columns=['Order Month'])
OrderMonthDim['Order_Month_Id'] = OrderMonthDim.index
OrderMonthDim
```

```python
# Creating Dimension table for Order Year
OrderYearDim = fact_table['Order Year'].unique()
OrderYearDim = pd.DataFrame(OrderYearDim, columns=['Order Year'])
OrderYearDim['Order_Year_Id'] = OrderYearDim.index
OrderYearDim
```

```python
# Creating Dimension table for Shipping Month
ShippingMonthDim = fact_table['Shipping Month'].unique()
ShippingMonthDim = pd.DataFrame(ShippingMonthDim, columns=['Shipping Month'])
ShippingMonthDim['Shipping_Month_Id'] = ShippingMonthDim.index
ShippingMonthDim
```

```python
# Creating Dimension Table for Shipping Year
ShippingYearDim = fact_table['Shipping Year'].unique()
ShippingYearDim = pd.DataFrame(ShippingYearDim, columns=['Shipping Year'])
ShippingYearDim['Shipping_Year_Id'] = ShippingYearDim.index
ShippingYearDim
```

```python
fact_table.columns
```

```python
#Dropping Columns from fact table {Dimension tables headers}
fact_table.drop(['Market', 'Shipping Mode', 'Customer Country', 'Customer City', 'Customer Segment'
                 , 'Type', 'Department Name', 'Order City', 'Order Country', 'Category Name', 'shipping date (DateOrders)',
                   'Order Day', 'Shipping Year', 'Shipping Month', 'Shipping Day',
                     'Order Region', 'Order State'
                     ], axis=1, inplace=True)
```
